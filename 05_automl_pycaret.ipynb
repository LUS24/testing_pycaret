```python
import pandas as pd
from pycaret.time_series import *
from pycaret.utils import enable_colab
enable_colab()
```

## AutoML Pycaret con datos agrupados por semana

```python
data = pd.read_csv("https://raw.githubusercontent.com/LUS24/testing_pycaret/main/data/serie_completa.csv")
data.fecha = pd.to_datetime(data.fecha)
data = data.groupby([pd.Grouper(key='fecha', freq='1W')]).mean()
data = data.loc['2020-01-01':] # Data subset to reduce training time
```

```python
data
```

```python
s = setup(data, fh = 3, fold = 5, session_id = 123, target="ice_conc")
```

```python
best =  compare_models()
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 3}, save=True)
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 4}, save=True)
```

```python
# finalize model
final_best = finalize_model(best)
predict_model(best, fh = 4)
```
